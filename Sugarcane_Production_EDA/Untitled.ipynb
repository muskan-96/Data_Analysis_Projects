```python
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')
f = 2*x + 1

p = plot(f, show=False)   # create the plot
p[0].line_color = 'blue'  # optional
p.xlabel = 'x'
p.ylabel = 'y'
p.show()                  # now display with axis numbers

```

```python
import numpy as np

# Create x values from -10 to 10
x_values = np.arange(-10, 11)  # -10, -9, ..., 10

# Function: f(x) = x^2 + 1
y_values = []

for x in x_values:
    y = x**2 + 1
    y_values.append(y)

print("x values:", x_values)
print("y values:", y_values)
import matplotlib.pyplot as plt

plt.plot(x_values, y_values, marker='o')  # 'o' will show points
plt.xlabel("x")
plt.ylabel("f(x) = x^2 + 1")
plt.title("Plot of f(x) = x^2 + 1")
plt.grid(True)
plt.show()

```
