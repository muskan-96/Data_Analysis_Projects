```python
import pandas as pd
df=pd.read_csv('BlackFriday.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df
```

```python
df.groupby('Age').size().plot(kind='bar',figsize=(10,6),title='Purchase Distribution by Age')
```

```python
for i in df['Age'].unique():
    print(i,":",df[df['Age']== i]['Product_ID'].nunique())
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
data
```

```python
data.plot.bar(x = 'Age', figsize = (8,6))
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (12, 6), title = 'Amount Spend by Age')
```

```python
df.groupby('Age')['Purchase'].mean().plot(
    kind='pie', figsize=(6,6),autopct='%0.1f', title='Average Purchase by Age Group'
)
```

```python
df.groupby('Marital_Status').size().plot(
    kind='pie', figsize=(6,6),autopct='%0.1f')
```

```python
df.groupby('Gender').size().plot(
    kind='pie', figsize=(6,6),autopct='%0.1f')
```
