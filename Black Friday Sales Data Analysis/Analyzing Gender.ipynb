```python
del df['Product_Category_2']
del df['Product_Category_3']
```

## Naive Approach

```python
data=pd.DataFrame({'Ratio':[len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]},index=["Male","Female"])
```

```python
data
```

```python
data.plot.pie(y = 'Ratio',figsize=(6,6),autopct="%.2f")
```

## Using column

```python
df.groupby('Gender').size().plot(kind='pie',autopct="%.2f",title='Gender Ratio',figsize=(6,6))
```

```python
df.groupby('Gender').size().plot(kind='bar',title='Gender Ratio',figsize=(6,6))
```

```python
df.groupby('Gender').size()
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind = 'pie', autopct = "%0.1f")
```
